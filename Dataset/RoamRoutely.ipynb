```python
# Import necessary libraries (train, validation and test)
import numpy as np
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
destination = pd.read_csv('/content/dataset.csv')
```

```python
def match_destination_hotel(input_Destination, budget):

    # Exact string matching for destination
    exact_destination_matches = destination[destination['District'].fillna('').str.lower() == input_Destination.lower()]

    # Filter destinations by budget range
    if not exact_destination_matches.empty:
        budget_matches = exact_destination_matches[
            (exact_destination_matches['Hotel_Budget'] >= int(budget) - 20) &
            (exact_destination_matches['Hotel_Budget'] <= int(budget) + 20)
        ]
        if not budget_matches.empty:
            return budget_matches.to_dict('records')

    # Fuzzy string matching for destination
    fuzzy_matches = [(index, cn, fuzz.partial_ratio(input_Destination.lower(), str(cn).lower())) for index, cn in enumerate(destination['District'])]
    fuzzy_matches = sorted(fuzzy_matches, key=lambda x: x[2], reverse=True)

    best_match_index, best_match_destination, score = fuzzy_matches[0]
    if score > 80:  # You can adjust this threshold as needed
        best_match_budget = destination.iloc[best_match_index]['Hotel_Budget']
        if abs(int(budget) - best_match_budget) <= 20:
            return [destination.iloc[best_match_index].to_dict()]

    return None  # No match found



def match_destination_restaurants(input_Destination, budget):

    # Exact string matching for destination
    exact_destination_matches = destination[destination['District'].fillna('').str.lower() == input_Destination.lower()]

    # Filter destinations by budget range
    if not exact_destination_matches.empty:
        budget_matches = exact_destination_matches[
            (exact_destination_matches['Restaurant_Budget'] >= int(budget) - 20) &
            (exact_destination_matches['Restaurant_Budget'] <= int(budget) + 20)
        ]
        if not budget_matches.empty:
            return budget_matches.to_dict('records')

    # Fuzzy string matching for destination
    fuzzy_matches = [(index, cn, fuzz.partial_ratio(input_Destination.lower(), str(cn).lower())) for index, cn in enumerate(destination['District'])]
    fuzzy_matches = sorted(fuzzy_matches, key=lambda x: x[2], reverse=True)

    best_match_index, best_match_destination, score = fuzzy_matches[0]
    if score > 80:  # Threshold
        best_match_budget = destination.iloc[best_match_index]['Restaurant_Budget']
        if abs(int(budget) - best_match_budget) <= 20:
            return [destination.iloc[best_match_index].to_dict()]

    return None  # No match found


# Example
input_destination = input("Enter the location: ").strip()
budget = input("Enter budget: ").strip()
time_frame = input("Enter the number of days of the trip: ")
people_count = input("Enter the number of people traveling: ")
# vaca_type = input("Enter the type of vacation: ")
print()

matched_hotel_destinations = match_destination_hotel(input_destination, budget)
matched_restaurant_destinations = match_destination_restaurants(input_destination, budget)

if matched_hotel_destinations:
    print("Hotels found in given location", input_destination + ":", len(matched_hotel_destinations))
    print()
    for matched_dest in matched_hotel_destinations:
      print(matched_dest["Hotels"])
else:
  print()
  print("No hotels found for the given location and budget.")

print()

if matched_restaurant_destinations:
    print("Restaurants found in given location", input_destination + ":", len(matched_restaurant_destinations))
    print()
    for matched_dest in matched_restaurant_destinations:
       print(matched_dest["Restaurants"])
else:
  print()
  print("No restaurants found for the given location and budget.")
```
